```python
def bin2dec(arr1):
    x = arr1.shape[0]
    decimal = 0
    for i in range(x):
        decimal = decimal + (2**i)*(arr1[6-i])
    return decimal

def dec2bin(pixel):
    bin_value = []
    if (pixel>1) :
        bin_value = dec2bin(pixel//2)
    bin_value.append(pixel%2)
    return bin_value
```
